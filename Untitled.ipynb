```python
import os
import json
import re
import random
```

```python
fulllist = []

for root, dirs, files in os.walk('json'):
    for file in files:
        with open(os.path.join(root, file), 'r', encoding = 'utf-8') as f:
            d = json.loads(f.read())
        for l in d:
            if l['content'] == None:
                continue
            l['content'] = re.sub(r'<.*?>','',l['content'])
            l['content'] = l['content'].replace('\u3000', '')
            
            fulllist.append(l)
            
        
```

```python
print(len(fulllist))
```

```python
def getname(fulllist):
    sentence_list = []
    #print('start')
    while len(sentence_list) < 2:
        article = random.choice(fulllist)
        content = re.sub(r'[（(].+[)）]', '', article['content'])
        sentence_list = re.split(r'[。？！\s]', content)
        #print(sentence_list)
    sentence = ''
    while len(sentence) < 3:
        sentence = random.choice(sentence_list)
    #print(sentence)
    b_list = range(0,len(sentence))
    #print(b_list)
    name = ''
    while len(name) != 2:
        l = random.sample(b_list,2)
        l = sorted(l)
        for i in l:
            c = sentence[i]
            if c.isalpha():
                name += c
        if len(name) != 2:
            name=''

    return name,l,sentence,article['book'],article['title'],article['author']
```

```python
for i in range(0,10000):
    print(getname(fulllist))
```
